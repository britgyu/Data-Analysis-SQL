### ¿Cuál es el año que registra más accidentes (LESIONES + FALLECIDOS) ?

```python
%%sql

SELECT ANIO, SUM(LESIONADOS) AS Lesionados, SUM(FALLECIDOS) AS Fallecidos, SUM(LESIONADOS) + SUM(FALLECIDOS) AS Total
FROM siniestros
GROUP BY ANIO
ORDER BY Total DESC
```

El año que más personas fallecidas reportó fue 2017

### Top 10 de automoviles involucrados

```python
%%sql

SELECT TIPO_DE_VEHICULO_1, COUNT(*) AS Total, ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(),2) AS Porcentaje 
FROM Siniestros
GROUP BY TIPO_DE_VEHICULO_1
ORDER BY Total DESC
LIMIT 10
```

Los automóviles y bicicletas representan aproximadamente el 50% de los accidentes
